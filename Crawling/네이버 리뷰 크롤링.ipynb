```python
driver = webdriver.Chrome()
# driver = webdriver.Chrome('chromedriver', options=options)
url = 'https://m.place.naver.com/place/1960638614/review/visitor?reviewItem=0'
driver.get(url)

data = pd.DataFrame(data=[], columns=['별점','리뷰'])

while True:
        try:
            time.sleep(1)
            driver.find_element_by_tag_name('body').send_keys(Keys.END)
            
            time.sleep(3)
            driver.find_element_by_xpath('//*[@id="app-root"]/div/div/div/div[6]/div[4]/div[3]/div[2]/a').click()
            time.sleep(3)
            
            driver.find_element_by_tag_name('body').send_keys(Keys.END)
            time.sleep(1)
            
        except NoSuchElementException:
            print('-더보기 버튼 모두 클릭 완료-')
            break
```

```python
# 수집할 정보들
rating_list = [] # 평점
user_review_id = {} # 유저의 id 
review_json = {} # 리뷰
image_json = {} # 이미지

# 파싱
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')

waterpark_name = "신화워터파크"

user_review_id[waterpark_name] = {}
review_json[waterpark_name] = {}

try:
    ul_review = soup.find('ul', attrs={'class':'_1jVSG'})
    one_review = ul_review.find_all('li', attrs = {'class':'_3FaRE'})
    review_num = len(one_review)
    print('리뷰 총 개수 :'+str(review_num))
    
    
    for i in range(len(one_review)):
        
        # rating(별점)
        try:
            rating = one_review[i].find('div', attrs = {'class':'_3-LAD'}).text
            print('rating =' + rating)
        except NoSuchElementException:
            rating=""
            print('rating = NaN')
            
        # reivew 내용
        try:
            review_content = one_review[i].find('span', attrs = {'class': 'WoYOw'}).text
        
        except:
            review_content = ""
        print('리뷰 내용:' + review_content)
        
        #review_json
        review_json[waterpark_name] = review_content
        naver_review = {waterpark_name, rating, review_content}
        rating_list.append(naver_review)
        
        tmp =[]
        tmp.append(rating)
        tmp.append(review_content)
        tmp = pd.DataFrame(data=[tmp], columns=data.columns)
        data = pd.concat([data,tmp])

except NoSuchElementException:
            none_review = "네이버 리뷰 없음"
            print(none_review)
            review_num = 0

print('\n')
```

```python
data.to_csv('신화워터파크_크롤링.csv', header= False, index =False)
```
